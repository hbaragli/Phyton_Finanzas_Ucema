##  Generar la lista de retornos

```python
%%timeit
n = 5000
ret = np.random.normal(0,2,n)
```

```python
%%timeit
n = 5000
ret = [ random.normalvariate(mu=0, sigma=2)  for i in range(n)]
```

```python
n = 5000
ret = np.random.normal(0,2,n)
ret
```

```python
ret.std()
```

$$  \Large  \sigma = \sqrt {\frac{1}{n} \sum_{i=1}^{n} (\mu - x_i)^2}  $$

```python
suma = 0
mu = ret.mean()

for xi in ret:
  suma += (mu - xi)**2

(suma/n)**0.5
```
